```python
import pandas as pd
```

```python
student=pd.read_csv('https://raw.githubusercontent.com/annapoorna-a-k/STUDENT-GRADE-PREDICTION-using-Linear-Regression/main/student-mat.csv')
```

```python
student.columns
```

```python
student.describe()
```

```python
y=student['G3']
```

```python
x=student[['age',	'Medu',	'Fedu',	'traveltime',	'studytime',	'failures',	'freetime',	'goout',	'health',	'absences',	'G1',	'G2']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=2529)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```
